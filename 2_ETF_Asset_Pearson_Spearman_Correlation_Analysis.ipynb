```python
df_etf_final = pd.read_csv('/content/gdrive/MyDrive/NH 공모전/data/df_etf_final.csv')
```

# 피어슨/스피어만 상관계수 구하기

```python
# None 값을 NaN으로 변환
df_etf_final['금_증감률'] = pd.to_numeric(df_etf_final['금_증감률'], errors='coerce')
df_etf_final['코인_증감률'] = pd.to_numeric(df_etf_final['코인_증감률'], errors='coerce')
df_etf_final['환율_증감률'] = pd.to_numeric(df_etf_final['환율_증감률'], errors='coerce')
df_etf_final['yield_rate'] = pd.to_numeric(df_etf_final['yield_rate'], errors='coerce')

def calculate_correlations(df, column, yield_rate='yield_rate'):
    return df.groupby('tck_iem_cd').apply(
        lambda x: pd.Series({
            f'{column}_피어슨': x[column].corr(x[yield_rate]),
            f'{column}_스피어만': x[column].corr(x[yield_rate], method='spearman')
        })
    ).reset_index()

# 각 종목별로 상관계수 계산
gold_corr_df = calculate_correlations(df_etf_final, '금_증감률')
coin_corr_df = calculate_correlations(df_etf_final, '코인_증감률')
exchange_corr_df = calculate_correlations(df_etf_final, '환율_증감률')
```

```python
print(gold_corr_df)
print(coin_corr_df)
print(exchange_corr_df)
```

```python
# 첫 번째와 두 번째 데이터프레임 병합
corr_df = pd.merge(gold_corr_df, coin_corr_df, on='tck_iem_cd', how='outer')

# 병합된 데이터프레임과 세 번째 데이터프레임 병합
corr_df = pd.merge(corr_df, exchange_corr_df, on='tck_iem_cd', how='outer')

# 결과 출력
corr_df
```

# 상관계수 0.3
* 상관관계가 둘 다 0.3 이상이거나 이하이면 1로 표시
* 라벨이 1인 종목명만 추출

```python
columns = ['금', '코인', '환율']

for i in columns:
  corr_df[i + '_양_상관관계'] = ((corr_df[i + '_증감률_피어슨'] >= 0.3) & (corr_df[i + '_증감률_스피어만'] >= 0.3)).astype(int)
  corr_df[i + '_음_상관관계'] = ((corr_df[i + '_증감률_피어슨'] <= -0.3) & (corr_df[i + '_증감률_스피어만'] <= -0.3)).astype(int)

corr_df
```

```python
# 종목별 상관관계에 따라 필터링 및 리스트 생성
def get_related_names(corr_df, columns):
    results = {}
    for col in columns:
        pos_related = corr_df[corr_df[f'{col}_양_상관관계'] == 1]['tck_iem_cd'].tolist()
        neg_related = corr_df[corr_df[f'{col}_음_상관관계'] == 1]['tck_iem_cd'].tolist()
        results[f'{col}_양_3'] = pos_related
        results[f'{col}_음_3'] = neg_related
    return results

# 상관관계 분석할 열 목록
related_names = get_related_names(corr_df, columns)

# 결과 출력
for key, names in related_names.items():
    print(names)
    print(f'{key} 0.3 ETF 개수:', len(names))
```

```python
# 결과를 데이터프레임으로 변환
corr_final_df = pd.DataFrame(dict([(key, pd.Series(names)) for key, names in related_names.items()]))
corr_final_df
```
